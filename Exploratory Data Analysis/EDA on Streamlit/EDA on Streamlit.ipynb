```python
# Assuming your DataFrame is named 'data'
null_cnt = data.isnull().sum()
null_col = null_cnt != 0

# Filter the DataFrame to include only columns with null values
filtered_data = data.loc[:, null_col]

# Find numeric columns in filtered_data
numeric_columns = filtered_data.select_dtypes(include=[int, float]).columns.tolist()

# Find non-numeric columns in filtered_data
non_numeric_columns = filtered_data.select_dtypes(exclude=[int, float]).columns.tolist()

print("Numeric Columns:", numeric_columns)
print("Non-Numeric Columns:", non_numeric_columns)
```

```python
# Replace null values with median/mode as appropriate
median = data[['GST(in%)','Weight']].median()
data[['GST(in%)','Weight']] = data[['GST(in%)','Weight']].fillna(median)
print(median)

mode = data[['Manufacturer','Provider','Client','GST Reg No','Patient','Species','Breed','Gender']].mode().iloc[0]
data[['Manufacturer','Provider','Client','GST Reg No','Patient','Species','Breed','Gender']].fillna(mode)
print(mode)
```

```python
# Encode non-numeric features
categorical_columns = ['Manufacturer', 'Provider', 'Client', 'GST Reg No', 'Patient', 'Species', 'Breed', 'Gender']
labelencoder = LabelEncoder()
data = pd.get_dummies(data, columns=categorical_columns)
print(data.head())
```

```python
# Correlation between two continuous variables
print(data['Qty'].corr(data['Cost']))
```

```python
# Correlation between two categorical variables
# Create a contingency table
contingency_table = pd.crosstab(data['Services and Goods'], data['Plan Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output the results
print("Chi-square value:", chi2)
print("P-value:", p)
```
